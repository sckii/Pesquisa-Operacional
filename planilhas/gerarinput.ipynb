```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
onlyfiles = [f for f in listdir("./") if isfile(join("./", f))]
onlyfiles.remove('gerarinput.ipynb')
onlyfiles.remove('input.csv')
onlyfiles
```

```python
produtosDF = []
for path in onlyfiles:
  nome_produto = path.split("Ficha Formatada - ")[1].split(".csv")[0]
  print(nome_produto)
  df = pd.read_csv(path)
  df.insert(0, 'Produto', nome_produto)
  produtosDF.append(df)
```

```python
input_dataframe = pd.DataFrame()

for produtoDF in produtosDF:
  input_dataframe = pd.concat([input_dataframe, produtoDF])

input_dataframe
```

```python
pd.DataFrame(input_dataframe).to_csv("input.csv", sep=',', encoding='utf-8', index=False, header=True)
```
